**Rotated images**

```python
def rotation_image(image):

    rotated_images = []
    fig = plt.figure()
    plt.subplot(1,4,1)
    plt.title('Original image')
    plt.imshow(image)

    for i in range (3):

        rotated_image = image.rotate(90*(i+1))
        rotated_images.append(rotated_image)

        plt.subplot(1,4,(i+2))
        plt.title('Rotated image')
        plt.imshow(rotated_image)
    
    return rotated_images
```

```python
image = Image.open(r"training/images/satImage_001.png")

rotated_images = rotation_image(image)

for i in range(len(rotated_images)):
    rotated_image = rotated_images[i]
    rotated_image.show()
```

**Brightness change**

```python
def brightness_image(image):
    image_enhanced = ImageEnhance.Brightness(image)
    brightness_images = []
    fig = plt.figure()
    plt.subplot(1,11,1)
    plt.title('Original image')
    plt.imshow(image)

    for i in range (10):

        factor = (i+1)*0.1
        enhanced_output = image_enhanced.enhance(factor)
        #enhanced_output.show()
        brightness_images.append(enhanced_output)

        plt.subplot(1,11,i+1)
        plt.title('Brightness image of factor'+str(i))
        plt.imshow(enhanced_output)

    return brightness_images

```

```python
image = Image.open(r"training/images/satImage_001.png")

brightness_images = brightness_image(image)

for i in range(len(brightness_images)):
    brightness_image_i = brightness_images[i]
    brightness_image_i.show()
```
