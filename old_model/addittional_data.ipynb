**Random Perspective**

```python
image = Image.open(r"training/images/satImage_001.png")

perspective_transformer = v2.RandomPerspective(distortion_scale=0.6, p=1.0)

perspective_imgs = [perspective_transformer(image) for _ in range(4)]

for i in range (len(perspective_imgs)):
    perspective_imgs[i] = v2.Resize(400)(perspective_imgs[i])
    perspective_imgs[i].show()
```

```python
# Paths to your folders
image_folder = r"training/images"
groundtruth_folder = r"training/groundtruth"
additional_groundtruth = r"training/additional_groundtruth"
additional_images = r"training/additional_images"

sizes = [30, 50, 100, (400, 400)]  # Different sizes for cropping and resizing

# the transform : perspective transformer
perspective_transformer = v2.RandomPerspective(distortion_scale=0.6, p=1.0)

#loop along the original images
for i in range(1, 101):

    #path of the original image
    image_name = f"satImage_{i:03}.png"
    image_path = os.path.join(image_folder, image_name)
    
    # Open the original image
    image = Image.open(image_path)

    # Apply the perspective transform and save 4 different perspectives
    perspective_imgs = [perspective_transformer(image) for _ in range(4)]

    for j, perspective_img in enumerate(perspective_imgs):
        # Resize each perspective-transformed image to 400x400
        resized_perspective_img = v2.Resize(400)(perspective_img)

        # Save the transformed image
        input_filename = f"satImage_{i:03}_perspective_{j+1}.png"  # Unique name for each perspective
        input_path = os.path.join(additional_images, input_filename)
        resized_perspective_img.save(input_path)
        print(f"Saved perspective image {j+1} to {input_path}")

        # Load and process the corresponding groundtruth image
        groundtruth_name = f"satImage_{i:03}.png"  # Assume the groundtruth has the same name
        groundtruth_path = os.path.join(groundtruth_folder, groundtruth_name)
        
        groundtruth = Image.open(groundtruth_path)

        # Apply perspective transform to the ground truth image as well (if desired)
        perspective_groundtruth = perspective_transformer(groundtruth)

        # Resize the groundtruth perspective to 400x400
        resized_perspective_groundtruth = v2.Resize(400)(perspective_groundtruth)

        # Save the perspective-transformed groundtruth image
        output_filename = f"satImage_{i:03}_perspective_{j+1}.png"  # Same name as the input image but for groundtruth
        output_path = os.path.join(additional_groundtruth, output_filename)
        resized_perspective_groundtruth.save(output_path)
        print(f"Saved perspective groundtruth {j+1} to {output_path}")
```

**Random Rotation**

```python
image = Image.open(r"training/images/satImage_001.png")

rotater = v2.RandomRotation(degrees=(0, 180))

rotated_imgs = [rotater(image) for _ in range(4)]

for i in range (len(rotated_imgs)):
    rotated_imgs[i] = v2.Resize(400)(rotated_imgs[i])
    rotated_imgs[i].show()
```

```python
# Paths to your folders
image_folder = r"training/images"
groundtruth_folder = r"training/groundtruth"
additional_groundtruth = r"training/additional_groundtruth"
additional_images = r"training/additional_images"

sizes = [30, 50, 100, (400, 400)]  # Different sizes for cropping and resizing

# transformer : random rotation
rotater = v2.RandomRotation(degrees=(0, 180))

#loop along the original images
for i in range(1, 101):
    
    #path of the original image
    image_name = f"satImage_{i:03}.png"
    image_path = os.path.join(image_folder, image_name)
    
    # Open the original image
    image = Image.open(image_path)

    # Apply the random rotation and generate 4 different rotated versions
    rotated_imgs = [rotater(image) for _ in range(4)]

    for j, rotated_img in enumerate(rotated_imgs):
        # Resize each rotated image to 400x400
        resized_rotated_img = v2.Resize(400)(rotated_img)

        # Save the rotated image
        input_filename = f"satImage_{i:03}_rotated_{j+1}.png"  # Unique name for each rotation
        input_path = os.path.join(additional_images, input_filename)
        resized_rotated_img.save(input_path)
        print(f"Saved rotated image {j+1} to {input_path}")

        # Load and process the corresponding groundtruth image
        groundtruth_name = f"satImage_{i:03}.png"  # Assume the groundtruth has the same name
        groundtruth_path = os.path.join(groundtruth_folder, groundtruth_name)
        
        groundtruth = Image.open(groundtruth_path)

        # Apply the same random rotation to the ground truth image
        rotated_groundtruth = rotater(groundtruth)

        # Resize the rotated groundtruth to 400x400
        resized_rotated_groundtruth = v2.Resize(400)(rotated_groundtruth)

        # Save the rotated groundtruth image
        output_filename = f"satImage_{i:03}_rotated_{j+1}.png"  # Same name as the input image but for groundtruth
        output_path = os.path.join(additional_groundtruth, output_filename)
        resized_rotated_groundtruth.save(output_path)
        print(f"Saved rotated groundtruth {j+1} to {output_path}")
```

**Random Affine**

```python
image = Image.open(r"training/images/satImage_001.png")

affine_transfomer = v2.RandomAffine(degrees=(30, 70), translate=(0.1, 0.3), scale=(0.5, 0.75))

affine_imgs = [affine_transfomer(image) for _ in range(4)]

for i in range (len(affine_imgs)):
    affine_imgs[i] = v2.Resize(400)(affine_imgs[i])
    affine_imgs[i].show()
```

```python
# Paths to your folders
image_folder = r"training/images"
groundtruth_folder = r"training/groundtruth"
additional_groundtruth = r"training/additional_groundtruth"
additional_images = r"training/additional_images"

sizes = [30, 50, 100, (400, 400)]

# Random Affine Transformer with degrees, translation, and scaling
affine_transfomer = v2.RandomAffine(degrees=(30, 70), translate=(0.1, 0.3), scale=(0.5, 0.75))

#loop along the original images
for i in range(1, 101):
    
    #path of the original image
    image_name = f"satImage_{i:03}.png"
    image_path = os.path.join(image_folder, image_name)
    
    # Open the original image
    image = Image.open(image_path)

    # Apply the random affine transformation and generate 4 different transformed versions
    affine_imgs = [affine_transfomer(image) for _ in range(4)]

    for j, affine_img in enumerate(affine_imgs):
        # Resize each affine-transformed image to 400x400
        resized_affine_img = v2.Resize(400)(affine_img)

        # Save the affine transformed image
        input_filename = f"satImage_{i:03}_affine_{j+1}.png"  # Unique name for each affine transformation
        input_path = os.path.join(additional_images, input_filename)
        resized_affine_img.save(input_path)
        print(f"Saved affine image {j+1} to {input_path}")

        # Load and process the corresponding groundtruth image
        groundtruth_name = f"satImage_{i:03}.png"  # Assume the groundtruth has the same name
        groundtruth_path = os.path.join(groundtruth_folder, groundtruth_name)
        
        groundtruth = Image.open(groundtruth_path)

        # Apply the same affine transformation to the ground truth image
        affine_groundtruth = affine_transfomer(groundtruth)

        # Resize the affine-transformed groundtruth to 400x400
        resized_affine_groundtruth = v2.Resize(400)(affine_groundtruth)

        # Save the affine-transformed groundtruth image
        output_filename = f"satImage_{i:03}_affine_{j+1}.png"  # Same name as the input image but for groundtruth
        output_path = os.path.join(additional_groundtruth, output_filename)
        resized_affine_groundtruth.save(output_path)
        print(f"Saved affine groundtruth {j+1} to {output_path}")
```

**Random Crop**

```python
image = Image.open(r"training/images/satImage_001.png")

cropper = v2.RandomCrop(size=(128, 128))

crops = [cropper(image) for _ in range(4)]

for i in range (len(crops)):
    crops[i] = v2.Resize(400)(crops[i])
    crops[i].show()
```

```python
# Paths to your folders
image_folder = r"training/images"
groundtruth_folder = r"training/groundtruth"
additional_groundtruth = r"training/additional_groundtruth"
additional_images = r"training/additional_images"

sizes = [30, 50, 100, (400, 400)]

# Random Crop Transformer with size=(128, 128)
cropper = v2.RandomCrop(size=(128, 128))

#loop along the original images
for i in range(1, 101):

    #path of the original image
    image_name = f"satImage_{i:03}.png"
    image_path = os.path.join(image_folder, image_name)
    
    # Open the original image
    image = Image.open(image_path)

    # Apply the random crop transformation and generate 4 different cropped versions
    crops = [cropper(image) for _ in range(4)]

    for j, crop in enumerate(crops):
        # Resize each cropped image to 400x400
        resized_crop = v2.Resize(400)(crop)

        # Save the cropped image
        input_filename = f"satImage_{i:03}_crop_{j+1}.png"  # Unique name for each cropped transformation
        input_path = os.path.join(additional_images, input_filename)
        resized_crop.save(input_path)
        print(f"Saved cropped image {j+1} to {input_path}")

        # Load and process the corresponding groundtruth image
        groundtruth_name = f"satImage_{i:03}.png"  # Assume the groundtruth has the same name
        groundtruth_path = os.path.join(groundtruth_folder, groundtruth_name)
        
        groundtruth = Image.open(groundtruth_path)

        # Apply the same random crop transformation to the ground truth image
        crop_groundtruth = cropper(groundtruth)

        # Resize the cropped groundtruth to 400x400
        resized_crop_groundtruth = v2.Resize(400)(crop_groundtruth)

        # Save the cropped groundtruth image
        output_filename = f"satImage_{i:03}_crop_{j+1}.png"  # Same name as the input image but for groundtruth
        output_path = os.path.join(additional_groundtruth, output_filename)
        resized_crop_groundtruth.save(output_path)
        print(f"Saved cropped groundtruth {j+1} to {output_path}")
```
